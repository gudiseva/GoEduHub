```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/gudiseva/GoEduHub/main/student_data.csv")
```

```python
df
```

```python
df.fillna(0)
```

```python
new_df = df.fillna(
    {'CS' : 0,
     'Math' : 0,
     'Name' : 'No Category'}
)
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") #index for row, column for columns
new_df
```

```python
new_df = df.fillna(method="ffill", limit=1)
new_df
```

```python
# new_df = df.fillna(method="ffill", limit=2, inplace=True) # <- Original DF is modified
df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all') # <- Rows having all the NaN will be deleted
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

```python
new_df = df.replace(np.NaN, 9999)
new_df
```

```python
new_df = df.replace(
    np.NaN, {
    'CS' : 999,
    'Math' : 0,
    'Name' : 'No Category'
})
new_df
```
